```python
df_tipo_preciomillon = inmuebles[['Tipo','Precio_Millon']]
df_tipo_preciomillon
```

```python
inmuebles_tipo_filtrado = df_tipo_preciomillon.drop(df_tipo_preciomillon[~df_tipo_preciomillon['Tipo'].isin(['Casa', 'Apartamento', 'Finca', 'Local'])].index)
inmuebles_tipo_filtrado
```

```python
plt.figure(figsize=(10,6))
grafica = sns.histplot(data=inmuebles_tipo_filtrado, x='Precio_Millon', kde=True, y='Tipo')
grafica.set_title('Distribucion valores precios inmuebles Bogota por tipo')
plt.xlim((50,3000))
plt.ylim((0,20))
plt.savefig('/content/drive/MyDrive/images/valor_inmuebles_por_tipo.png',format='png')
plt.show()
```

```python
plt.figure(figsize=(10,6))
grafica = sns.histplot(data=inmuebles_tipo_filtrado, x='Precio_Millon', kde=True, hue='Tipo')
grafica.set_title('Distribucion valores precios inmuebles Bogota por tipo')
plt.xlim((50,3000))
plt.ylim((0,100))
plt.savefig('/content/drive/MyDrive/images/valor_inmuebles_por_tipo.png',format='png')
plt.show()
```
